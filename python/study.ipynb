# Helicopter Dynamics


```python
import sympy as sp

# = Helicopter Properties =================================
blades_count_mr = sp.symbols("N_mr", integer=True, nonzero=True, positive=True)
power_mr, chord_mr, radius_mr, pitch_mr, lift_curve, profile_drag = sp.symbols(
    "P_MR, c_MR, R_MR, theta_MR, C_{L/\\alpha}, C_{D_0}",
    real=True,
    nonzero=True,
    positive=True,
)
weight = sp.symbols("W", real=True, nonzero=True, positive=True)

# = Rotor states ==========================================
rotation_speed_mr = sp.symbols("Omega_MR", real=True)

# = Air Properties ========================================
air_density = sp.symbols("rho", real=True, nonzero=True, positive=True)
```

Find the expression of thrust generate by the main rotor


```python
solidity_mr = blades_count_mr * chord_mr / (sp.pi * radius_mr)
inflow_mr = (
    (solidity_mr * lift_curve)
    / 16
    * (sp.sqrt(1 + (32 * pitch_mr) / (solidity_mr * lift_curve)) - 1)  # type: ignore
)
thrust_coeff_mr = (solidity_mr * lift_curve) / 2 * (pitch_mr / 3 - inflow_mr / 2)
thrust_mr = air_density * sp.pi * rotation_speed_mr**2 * radius_mr**4 * thrust_coeff_mr

print(r"Solidity $\sigma$")
display(solidity_mr)
print(r"Inflow $\lambda$:")
display(inflow_mr)
print(r"Thrust Coefficient $C_T$:")
display(thrust_coeff_mr)
print(r"Thrust $T$:")
display(thrust_mr)
```

Find the trim condition, in terms of main rotor pitch required, to have a vertical equilibrium


```python
vertical_trim_eq = sp.Eq(thrust_mr, weight)
trim_pitch_mr = sp.solve(vertical_trim_eq, pitch_mr)[0]

display(trim_pitch_mr)
```
